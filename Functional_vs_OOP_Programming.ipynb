Binary Search

```python
class Solution:
    def firstSearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        answer = -1

        while low <= high:
            mid = (low + high) // 2

            if arr[mid] == k:
                answer = mid
                high = mid - 1
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1

        return answer
        
```
